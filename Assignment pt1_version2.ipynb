### 1. Compute statistics on a Python list [5 points]
Given the following list
```
l = [12, 45, 12, 999, 10, 8, 76, 20, 10, 10, 7, 70, 17]
```
Compute the mean, the median and the standard deviation of the values in the list, using only built-in functionalities of python (no imported libraries, no NumPy or Pandas)

````python
# I am importing the statisitcs library to double-check my answers
import statistics

l = [12, 45, 12, 999, 10, 8, 76, 20, 10, 10, 7, 70, 17]

# Write your solution here

print('\033[1m Mean \033[0m')
print('-----------------------------------------------------------------------')
#dividing the sum of the list over its lenth to get the mean/average
l_mean = sum(l)/len(l)

print('The mean of list l is:', '\033[1m' , l_mean , '\033[0m')
print('Checking with the *mean()* function:', '\033[1m' , statistics.mean(l) , '\033[0m')
print()


print('\033[1m Median \033[0m')
print('-----------------------------------------------------------------------')
#updating list l by sorting all the values from least to greatest. This is needed to find the middle value
l = sorted(l)
#creating a variable for the length of list l which is 13
l_len = len(l)

#dividing the list length by 2 using integer divission since I want a whole number.
#the middle value is at position 7 in the list (in terms of length), 
#but since indexing starts at 0 I can keep this value and will obtain the middle value
l_median_index = l_len//2

#now I am printing the number that I obtained of the index
print('The median of list l is:', '\033[1m' ,l[l_median_index] , '\033[0m')
print('Checking with the *median()* function:', '\033[1m' , statistics.median(l) , '\033[0m')
print()

print('\033[1m Standard Deviation \033[0m')
print('-----------------------------------------------------------------------')

##the first step to find the standard deviation is to calculate the mean, which I have already done
#print(l_mean)

##the second step is to subtract each list value from the mean and to square that result
##I am creating a new list to store these values since I might need to original list in the future
w = []
i = 0
while i < l_len:
  w.append( (l[i] - l_mean)**2 )
  i = i + 1

##the third step is to then get the sum of the new values and divide it by the number in the list (the length)
##note: I am calculating the standard deviation of a sample so I am going to subtract the list length by 1 (which is 12)
##      if I was calculating the standard deviation of the population then I would divide by the full list length (which is 13)

sd_third_step = sum(w)/(l_len-1)

##the fourth step is to square the previous result
l_sample_sd = sd_third_step**0.5

print('The standard deviation of the sample of list l is:', '\033[1m' , l_sample_sd , '\033[0m')
print('Checking with the *stdev()* function:', '\033[1m' , statistics.stdev(l) , '\033[0m')
````

### Calculating the Key Statistics and finding the largest mean and smallest medain value of `sepal_width`

### Computing the correlation matrix/coefficient of `sepal_width` and `petal_length`

````python
### Compute the correlation coefficients here
y = 0 #this is dummy count variable that I used to ensure that I was printing the correct number of rows
s_p_width_length = {} #creating a dictionary that only contains the `sepal_width` and `petal_length` values 

for key in iris_dict_by_labels:
    sample_list = [] #this is a temporary list that will store all the data values of `sepal_width` and `petal_length`
    i = 0
    #print(dataset[1:3])
    #print(key)
    while i < 50: #this while loop makes sure to go through each dataset 50 times since each species has 50 rows of data
        sample_list = np.append(sample_list, iris_dict_by_labels[key][i][1:3])
        y = y+1
        i = i +1
    
    sample_list = sample_list.reshape(50, 2) #changing the list to match the proper array structure in the original dataset
    s_p_width_length[key] = sample_list #the i variable makes sure to iterate every row in the `iris_dict_by_labels` dictionary
        
    #print(y) #this ensures that the number of rows in each key is correct
    
### I am printing the dictionary that only contains the `sepal_width` and `petal_length` column values
#print(s_p_width_length) 

### solving for the covariance martrix
cov_mat = {
    key: dict(cov=np.cov(dataset.T, ddof=0)) for key, dataset in s_p_width_length.items()
}

#print(cov_mat)
#print('-------------------------------------')

### Compute the correlation matrix here
cor_mat = {
    key: dict(correlation_matrix=np.corrcoef(dataset.T)) for key, dataset in s_p_width_length.items()
}
#print(cor_mat)

#printing the correlation matrix in a easier to read format
for key in cor_mat:
    print('--------------------------------')
    print('#######',key,'#######')
    print('--------------------------------')
    for data in cor_mat[key]:
        print('###','\033[1m',data,'\033[0m','###')
        print(cor_mat[key][data])
    
````
