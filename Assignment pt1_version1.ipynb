### 1. Compute statistics on a Python list [5 points]
Given the following list
```
l = [12, 45, 12, 999, 10, 8, 76, 20, 10, 10, 7, 70, 17]
```
Compute the mean, the median and the standard deviation of the values in the list, using only built-in functionalities of python (no imported libraries, no NumPy or Pandas)

````python
# Note: I am importing these modules just to check my answers
from statistics import mean 
from statistics import median
from statistics import stdev
#import statistics

l = [12, 45, 12, 999, 10, 8, 76, 20, 10, 10, 7, 70, 17]

# Write your solution here
print('\033[1m This is the Mean: \033[0m')
print('-----------------------------------------------------------------------')
print('The mean of the list is:', sum(l)/len(l))
print('When I used the *mean()* function to check my answer I get', mean(l))
print()

print('\033[1m This is the Median: \033[0m')
print('-----------------------------------------------------------------------')
## I first had to sort the list from smallest to largest to get the middle value
l = sorted(l)

## I then had to get the list length which was 13 
#print(len(l))

## To get the middle value I have to divide the length by 2 and round the value
## note: when I do rounded division python goes to lowest number so 6.5 becomes 6, not 7. Because of the low rounding
##       I need to add 1 to properly round my result 'upward'. Now I did think about the possibility that dividing the 
##       result might give a value that might be 6.2 or 7.1 which would not need the upward rounding. However, list 
##       lengths come in whole numbers that can be odd or even and odd whole numbers always need to have upward rounding
##       when divided by 2 since the division result is always has a 0.5 in it (ex. 3/2 = 1.5,  15/2 = 7.5)
      
#The middle value of this list is 7. I now have to find the number on the 6th position/index of the list
#(I am saying 6 because indexing of lists start at 0)

#print( (len(l)//2) +1 )
print('The median of the list is:', l[len(l)//2])
print('When I used the *median()* function to check my answer I get',median(l))
print()

print('\033[1m This is the Standard Deviation: \033[0m')
print('-----------------------------------------------------------------------')
## The first step in finding the standard deviation is finding the mean which was previously calculated
## The mean is: 99.6923076923077
mean_l = sum(l)/len(l)

## The second step is to subtract each data point in the list by the mean and then to square the result
## I am creating a new list to store these new values
m = []
for x in range( len(l) ):
    m.append((l[x]-(mean_l))**2 )
    
## The third step is to find the sum of the values that were calculated in step 2. 
## The sum is 882890.7692307689
#print(sum(m))

## The 3rd step is to divide the sum with the number of data points in the list
## note: the number of data points here will be subtracted by one since I am calculating the standard deviation of the sample
## Since the list is small I assumed that I am dealing with a sample which is why I calculated the standard deviation of a sample
## The result that I got is 73574.23076923074
step_three_sd = sum(m)/(len(l)-1)

## The 4th and final step is to square root your previous result
print('The standard deviation (of the sample) of the list is: ', step_three_sd**0.5 )
print('When I used the *stdev()* function to check my answer I get:', stdev(l))
````

### 5. Create a new column from existing columns of a numpy array [5 points]

Create a new column for the flower volume V in `iris`, where volume is computing according to this formula (the assumption here is that the shape of the iris flower is approximately conical):

$$ Volume = \frac{\pi \times Sepal Length^2 \times Petal Length}{3}$$

````python
import math
# Input
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

'''iris = np.genfromtxt(url, delimiter=',', 
                     names=['sepal length', 'sepal width', 'petal length', 'petal width', 'species'],
                     dtype=[np.float32, np.float32, np.float32, np.float32, '|S15']
)'''
iris_2d = np.genfromtxt(url, delimiter=',', dtype='object')
# Write your solution here
iris_2d_list = [] #I am creating this list and will make it multi-dimensional. This list will contain the `volume` column and will be used to create the pandas dataframe in question 6

#iris_2d = iris_2d.reshape(150,6)
for x in iris_2d:    
    x[0:4] = x[0:4].astype(str).astype(float) #this converts all four length and width columns into a float type 
    x[4] = str(x[4], 'UTF-8') #this converts the species column into a string
    
    x = np.append(x, (math.pi* x[0]**2 * x[2])/3 ) #I am creating the `volume` column and use the append function to add the volume in each row of the `iris_2d` dataset      
    iris_2d_list.append(x) #this creates dynamic multidimensional list which allows the dataframe to formlater on
    

#print(iris_2d.dtype) #output: object
#print(iris_2d_list[:5]) #the output here confirms that the list formed is a multidimensional array
i = 0
for x in iris_2d_list:
    if i < 5:
        print(x)
    i = i +1
````
